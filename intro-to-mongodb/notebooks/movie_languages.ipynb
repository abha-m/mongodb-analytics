```python
pipeline = [
    {
        '$group' : {    # "group" stage : groups by a specified
                        # identifier expression, and appiles
                        # accumulator expression supplied to each
                        # group
            '_id': {"language" : "$language"}, # field-path identifier
            'count': {'$sum' : 1} # $sum is known as an accumulator
        }
    }
]

clear_output()
pprint.pprint(list(client.mflix.movies_initial.aggregate(pipeline)))
```
